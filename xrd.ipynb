```python
from pymatgen.ext.matproj import MPRester
from pprint import pprint
from pymatgen.analysis.diffraction.xrd import XRDCalculator
with MPRester("CBIGbbbqQQkuWWiae") as m:
    structure = m.get_structure_by_material_id("mp-169")
    structure.to(filename="C_mp-169.cif")
```

```python
x = XRDCalculator()
a = x.get_pattern(structure)
print(a)
```

```python
print("Premier pic du diffractogramme pour un rayonnement X incident Cu Kα :")
print(" ")
print("Indices hkl associés : ",a.hkls[0])
theta1 = (a.x)[0]
I1 = (a.y)[0]
print("Valeur de l'angle 2*\u03B8 (en degrés) : ",theta1)
print("Intensité du pic de Bragg (en %) : ",I1)
```

```python
print("Deuxième pic du diffractogramme pour un rayonnement X incident Cu Kα :")
print(" ")
print("Indices hkl associés : ",a.hkls[1])
print("                     : ",a.hkls[2])
theta2 = (a.x)[1]
I2 = (a.y)[1] + (a.y)[2]
print("Valeur de l'angle 2*\u03B8 (en degrés) : ",theta2)
print("Intensité du pic de Bragg (en %) : ",I2)
```

```python
print("Troisième pic du diffractogramme pour un rayonnement X incident Cu Kα :")
print(" ")
print("Indices hkl associés : ",a.hkls[3])
print("                     : ",a.hkls[4])
theta3 = (a.x)[3]
I3 = (a.y)[3] + (a.y)[4]
print("Valeur de l'angle 2*\u03B8 (en degrés) : ",theta3)
print("Intensité du pic de Bragg (en %) : ",I3)
```

```python
x.show_plot(structure,two_theta_range=(0, 90), annotate_peaks=False, ax=None, with_labels=True, fontsize=16)
```

Le graphe ci-dessus correspond au diffractogramme du carbone par des rayons X tels que calculés via la structure liée à l'identifiant du carbone mp-169. Le graphe ci-dessous est celui calculé automatiquement par le Materials Project. On remarque que les deux graphes correspondent parfaitement au niveau des angles. Il se trouve cependant que certains angles obtenus via le "DiffractionPattern" sont présents deux fois (dû à de minimes différences entre les deux). Il suffit dès lors de sommer les intensités obtenues pour les deux angles afin d'avoir le même résultat que sur le MP. Pour ce qui est des indices hkl, la différence entre les résultats obtenus plus haut et le MP vient du fait que nous avons les avons calculés pour la maille primitive alors que le MP les calcule pour la maille conventionnelle.
