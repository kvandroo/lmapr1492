```python
from pymatgen.ext.matproj import MPRester
from pprint import pprint
from pymatgen.analysis.diffraction.xrd import XRDCalculator
with MPRester("CBIGbbbqQQkuWWiae") as m:
    structure = m.get_structure_by_material_id("mp-169")
    structure.to(filename="C_mp-169.cif")
```

```python
x = XRDCalculator()
a = x.get_pattern(structure)
print(a)
```

```python
print("Premier pic du diffractogramme pour un rayonnement X incident Cu Kα :")
print(" ")
print("Indices hkl associés : ",a.hkls[0])
theta1 = (a.x)[0]
I1 = (a.y)[0]
print("Valeur de l'angle 2*\u03B8 (en degrés) : ",theta1)
print("Intensité du pic de Bragg (en %) : ",I1)
```

```python
print("Deuxième pic du diffractogramme pour un rayonnement X incident Cu Kα :")
print(" ")
print("Indices hkl associés : ",a.hkls[1])
print("                     : ",a.hkls[2])
theta2 = (a.x)[1]
I2 = (a.y)[1] + (a.y)[2]
print("Valeur de l'angle 2*\u03B8 (en degrés) : ",theta2)
print("Intensité du pic de Bragg (en %) : ",I2)
```

```python
print("Troisième pic du diffractogramme pour un rayonnement X incident Cu Kα :")
print(" ")
print("Indices hkl associés : ",a.hkls[3])
print("                     : ",a.hkls[4])
theta3 = (a.x)[3]
I3 = (a.y)[3] + (a.y)[4]
print("Valeur de l'angle 2*\u03B8 (en degrés) : ",theta3)
print("Intensité du pic de Bragg (en %) : ",I3)
```
